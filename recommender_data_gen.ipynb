```python
import pandas as pd
import os

DATA_DIR = os.path.join("data/anime_ori/")
df_info = pd.read_table(os.path.join(DATA_DIR, "anime_info.dat"), sep='\t')

df_info["genre"] = df_info["genre"].str.split(', ')
df_info = df_info.dropna()
cleaned_for_my_eyes = []
for i, l in enumerate(df_info["genre"].to_list()):
    if not any(elem in ["Hentai", "Yaoi", "Yuri"] for elem in l): #haha
        cleaned_for_my_eyes.append(df_info.iloc[i])

cleaned_for_my_eyes = pd.DataFrame(cleaned_for_my_eyes)
cleaned_for_my_eyes
```

```python
df_rating = pd.read_table(os.path.join(DATA_DIR, "anime_ratings.dat"), sep='\t')
df_rating
```

```python
df_rating = df_rating[df_rating["Anime_ID"].isin(cleaned_for_my_eyes["anime_ids"].unique())]
df_rating
```
